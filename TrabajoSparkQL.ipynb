```python
# Agrege acá el código para importar las librerias

# La libreria para "encontrar el sevicio" de Spark
import findspark
findspark.init()

# Librerias para "gestionar el servicio" de Spark
from pyspark import SparkConf, SparkContext
import pyspark
from pyspark.sql import SQLContext, SparkSession

# Creamos una aplicación Spark en el Servicio
# Tenga cuidado con las tildes o caracteres especiales en el nombre de la app
AppSpark = SparkConf().setAppName("Evaluacion iSofware")

# definimos un espacio o contexto para la App
ContextoSpark=SparkContext(conf=AppSpark)

# inicio una sesión en el espacio de la App
SesionSpark = SparkSession(ContextoSpark)

# inicio del espacio o contexto SQL
ContextoSql = SQLContext(sparkContext=ContextoSpark, sparkSession=SesionSpark)
```

```python
# 2. Cree dos tablas SparkSQL y almacene el csv en su correspondiente tabla.
# Observación: tenga especial cuidado con los encabezados de los archivos CSV.
# Usted puede considerar cambiar los encabezados de los CSV originales
# Tu código a continuación...

# Tabla asistencias
ContextoSql.sql("DROP TABLE IF EXISTS asistencia")
ContextoSql.sql("""
CREATE TABLE
asistencia (
    marca_temporal TIMESTAMP,
    usuario STRING,
    equipo_pertenece STRING,
    equipo_expone STRING    
    )
USING CSV
OPTIONS (header 'true',
         timestampFormat=\"yyyy/MM/dd HH:mm:ss\", 
         path 'Asistencia del Pitch (2021.04.13 ISoftware).csv')
"""
)

#Tabla evaluacion
ContextoSql.sql("DROP TABLE IF EXISTS evaluacion_read")
ContextoSql.sql("""

CREATE TABLE
evaluacion_read (
    marca_temporal TIMESTAMP,
    usuario STRING,
    equipo_evaluar STRING,
    introduccion INT,
    equipo INT,
    problemas INT,
    ventajas INT,
    solucion INT,
    prodcuto INT,
    traccion INT,
    mercado INT,
    competencia INT,
    modelo INT,
    inversion INT,
    contacto INT,
    expo_cordinacion INT,
    expo_claridad INT,
    expo_diapositivas INT,
    invertir STRING
    )
USING CSV
OPTIONS (path 'evaluacion.csv', 
         header 'true',
         timestampFormat=\"yyyy/MM/dd HH:mm:ss\",
         delimiter ";")
"""
)

#Obtenemos solo la ultima calificación hecha al equipo por el usuario
ContextoSql.sql("""
SELECT 
    MAX(marca_temporal) AS marca_temporal,
    usuario,
    equipo_evaluar 
FROM
    evaluacion_read
GROUP BY 
    usuario,equipo_evaluar

""").createOrReplaceTempView("evaluacion_distinct")

ContextoSql.sql("""
SELECT
    evaluacion_read.*
FROM
    evaluacion_distinct
LEFT JOIN
    evaluacion_read
    ON
    evaluacion_read.marca_temporal = evaluacion_distinct.marca_temporal
    AND evaluacion_read.usuario = evaluacion_distinct.usuario
    AND evaluacion_read.equipo_evaluar = evaluacion_distinct.equipo_evaluar

"""
).createOrReplaceTempView("evaluacion")
```

```python
# 2. Consulte el listado total de estudiantes (correos electrónicos) del 
# curso de Ingeniería de Software, ordenados alfabéticamente
# Tu código a continuación...

ContextoSql.sql("select distinct(usuario) from asistencia order by usuario").show(90)
```

```python
# 3. Consulte la cantidad de asistencias registradas por estudiante; además, la fecha y hora de la primera asistencia
# y la fecha y hora de la última asistencia
# Tu código a continuación...

ContextoSql.sql("""SELECT usuario,
                   COUNT(usuario) as asistencias,
                   MIN(marca_temporal) as min_fecha,
                   MAX(marca_temporal) as max_fecha
                   FROM asistencia
                   GROUP BY usuario""").show(90)
```

```python
# 4. Consulte el listado de estudiantes que asistieron a 2 presentaciones o menos (una).
# Tu código a continuación...

ContextoSql.sql("""SELECT usuario,
                   COUNT(equipo_expone) as asistencias
                   FROM asistencia
                   GROUP BY usuario
                   HAVING COUNT(equipo_expone) <=2 """).show()
```

```python
# 5. Consulte el listado de estudiantes que no asistieron a ninguna presentación.
# Tu código a continuación...

# Personas con asistencias
ContextoSql.sql(""" SELECT b.usuario AS evaluo,a.usuario AS asistio
                   FROM ( 
                   SELECT DISTINCT usuario FROM asistencia 
                   ) a
                   RIGHT JOIN  (
                   SELECT DISTINCT usuario FROM evaluacion 
                   ) b
                    on a.usuario=b.usuario
                    WHERE a.usuario is NULL
                   """).show()
```

```python
# 6. Consulte los integrantes por cada equipo al que pertenecen.
# Tu código a continuación...

ContextoSql.sql("""Select DISTINCT equipo_pertenece,
                          usuario
                from asistencia
                ORDER BY equipo_pertenece,usuario
                """).show(100)
```

```python
# 7. Consulte la cantidad de asistentes por presentación, sin considerar los asistentes que pertenecen 
# al equipo que realizó la presentación.
# Tu código a continuación...

ContextoSql.sql("""Select substring(equipo_expone,9) as expone,
                          count(distinct usuario) as cant_asistentes
                from asistencia
                where substring(equipo_expone,9) != substring(equipo_pertenece,15)
                group by expone
                order by expone
                """).show()
```

```python
# 8. Consutar cuáles integrantes evaluaron a su propio equipo. Estas evaluaciones no serán válidas, pues un
# integrante no puede evaluar a su propio equipo.
# Tu código a continuación...

ContextoSql.sql("""SELECT evaluacion.usuario,
                          evaluacion.equipo_evaluar,
                          equipo.equipo_pertenece
FROM evaluacion
LEFT JOIN (
Select DISTINCT substring(equipo_pertenece,15) as equipo_pertenece,usuario from asistencia
) equipo on (equipo.usuario=evaluacion.usuario)
WHERE equipo_evaluar=equipo_pertenece
""").show()
```

```python
# 9. Consultar la nota promedio por cada ítem (1 al 15), y la nota promedio total del cada equipo. Recuerde que no
# son válidas las evaluaciones realizadas por los miembros del mismo equipo.
# Tu código a continuación...


ContextoSql.sql("""SELECT evaluacion.equipo_evaluar,
                          ROUND(avg(introduccion),2) as cal_introduccion,
                          ROUND(avg(equipo),2) as cal_equipo,
                          ROUND(avg(problemas),2) as cal_problemas,
                          ROUND(avg(ventajas),2) as cal_ventajas,
                          ROUND(avg(solucion),2) as cal_solucion,
                          ROUND(avg(prodcuto),2) as cal_prodcuto,
                          ROUND(avg(traccion),2) as cal_traccion,
                          ROUND(avg(mercado),2) as cal_mercado,
                          ROUND(avg(competencia),2) as cal_competencia,
                          ROUND(avg(modelo),2) as cal_modelo,
                          ROUND(avg(inversion),2) as cal_inversion,
                          ROUND(avg(contacto),2) as cal_contacto,
                          ROUND(avg(expo_cordinacion),2) as cal_expo_cordinacion,
                          ROUND(avg(expo_claridad),2) as cal_expo_claridad,
                          ROUND(avg(expo_diapositivas),2) as cal_expo_diapositivas,
                          ROUND((avg(introduccion)+avg(equipo)+avg(problemas)+avg(ventajas)+avg(solucion)+
                          avg(prodcuto)+avg(traccion)+avg(mercado)+avg(competencia)+avg(modelo)+
                          avg(inversion)+avg(contacto)+avg(expo_cordinacion)+avg(expo_claridad)+avg(expo_diapositivas))/15,2) as final
FROM evaluacion
LEFT JOIN (
Select DISTINCT substring(equipo_pertenece,15) as equipo_pertenece,usuario from asistencia
) equipo on (equipo.usuario=evaluacion.usuario)
WHERE equipo_evaluar!=equipo_pertenece
GROUP BY equipo_evaluar
""").show()
```

```python
# 10. Consulte el mejor equipo evaluado por cada ítem (según nota promedio. En caso de empate mostrar todos los empatados) 
# y el mejor equipo según el promedio total.
# Tu código a continuación...
from pyspark.sql import *
from pyspark.sql.types import  StructType,StructField, StringType, IntegerType
import pyspark

ContextoSql.sql("""
SELECT evaluacion.equipo_evaluar,
                          ROUND(avg(introduccion),2) as cal_introduccion,
                          ROUND(avg(equipo),2) as cal_equipo,
                          ROUND(avg(problemas),2) as cal_problemas,
                          ROUND(avg(ventajas),2) as cal_ventajas,
                          ROUND(avg(solucion),2) as cal_solucion,
                          ROUND(avg(prodcuto),2) as cal_producto,
                          ROUND(avg(traccion),2) as cal_traccion,
                          ROUND(avg(mercado),2) as cal_mercado,
                          ROUND(avg(competencia),2) as cal_competencia,
                          ROUND(avg(modelo),2) as cal_modelo,
                          ROUND(avg(inversion),2) as cal_inversion,
                          ROUND(avg(contacto),2) as cal_contacto,
                          ROUND(avg(expo_cordinacion),2) as cal_expo_cordinacion,
                          ROUND(avg(expo_claridad),2) as cal_expo_claridad,
                          ROUND(avg(expo_diapositivas),2) as cal_expo_diapositivas,
                          ROUND((avg(introduccion)+avg(equipo)+avg(problemas)+avg(ventajas)+avg(solucion)+
                          avg(prodcuto)+avg(traccion)+avg(mercado)+avg(competencia)+avg(modelo)+
                          avg(inversion)+avg(contacto)+avg(expo_cordinacion)+avg(expo_claridad)+avg(expo_diapositivas))/15,2) as final
FROM evaluacion
LEFT JOIN (
Select DISTINCT substring(equipo_pertenece,15) as equipo_pertenece,usuario from asistencia
) equipo on (equipo.usuario=evaluacion.usuario)
WHERE equipo_evaluar!=equipo_pertenece
GROUP BY equipo_evaluar
""").createOrReplaceTempView("notas_equipo")

itemNames = ['cal_introduccion','cal_equipo','cal_problemas','cal_ventajas','cal_solucion','cal_producto','cal_traccion','cal_mercado','cal_competencia','cal_modelo','cal_inversion','cal_contacto','cal_expo_cordinacion','cal_expo_claridad','cal_expo_diapositivas','final']

schema = StructType([
    StructField("item", StringType(), True),
    StructField("equipo_evaluar", StringType(), True)
])

data = []
for itemName in itemNames:
    df_aux = ContextoSql.sql("""
    SELECT
        equipo_evaluar,""" + 
        itemName + """
    FROM
        notas_equipo
    WHERE
        """ + itemName + """ = (SELECT MAX(""" + itemName + """) FROM notas_equipo)
    """)
    for row in df_aux.rdd.collect():
        tupla = (itemName,row.equipo_evaluar)
        #print(tupla)
        data.append(tupla)
df = ContextoSql.createDataFrame(data=data,schema=schema)
df.show()
```

```python
# 11. Consulte el peor equipo evaluado por cada ítem (según nota promedio. En caso de empate mostrar todos los empatados) 
# y el peor equipo según el promedio total.
# Tu código a continuación...
schema = StructType([
    StructField("item", StringType(), True),
    StructField("equipo_evaluar", StringType(), True)
])

data = []
for itemName in itemNames:
    df_aux = ContextoSql.sql("""
    SELECT
        equipo_evaluar,""" + 
        itemName + """
    FROM
        notas_equipo
    WHERE
        """ + itemName + """ = (SELECT MIN(""" + itemName + """) FROM notas_equipo)
    """)
    for row in df_aux.rdd.collect():
        tupla = (itemName,row.equipo_evaluar)
        #print(tupla)
        data.append(tupla)
df = ContextoSql.createDataFrame(data=data,schema=schema)
df.show()
```

```python
# 12. Consulte la lista de estudiantes con la correspondiente nota obtenida en la presentación 
# (nota promedio total de la evaluación obtenida por el equipo)
# Tu código a continuación...

ContextoSql.sql("""  
with 
notas as (
SELECT evaluacion.equipo_evaluar,
                          ROUND(avg(introduccion),2) as cal_introduccion,
                          ROUND(avg(equipo),2) as cal_equipo,
                          ROUND(avg(problemas),2) as cal_problemas,
                          ROUND(avg(ventajas),2) as cal_ventajas,
                          ROUND(avg(solucion),2) as cal_solucion,
                          ROUND(avg(prodcuto),2) as cal_prodcuto,
                          ROUND(avg(traccion),2) as cal_traccion,
                          ROUND(avg(mercado),2) as cal_mercado,
                          ROUND(avg(competencia),2) as cal_competencia,
                          ROUND(avg(modelo),2) as cal_modelo,
                          ROUND(avg(inversion),2) as cal_inversion,
                          ROUND(avg(contacto),2) as cal_contacto,
                          ROUND(avg(expo_cordinacion),2) as cal_expo_cordinacion,
                          ROUND(avg(expo_claridad),2) as cal_expo_claridad,
                          ROUND(avg(expo_diapositivas),2) as cal_expo_diapositivas,
                          ROUND((avg(introduccion)+avg(equipo)+avg(problemas)+avg(ventajas)+avg(solucion)+
                          avg(prodcuto)+avg(traccion)+avg(mercado)+avg(competencia)+avg(modelo)+
                          avg(inversion)+avg(contacto)+avg(expo_cordinacion)+avg(expo_claridad)+avg(expo_diapositivas))/15,2) as final
FROM evaluacion
LEFT JOIN (
Select DISTINCT substring(equipo_pertenece,15) as equipo_pertenece,usuario from asistencia
) equipo on (equipo.usuario=evaluacion.usuario)
WHERE equipo_evaluar!=equipo_pertenece
GROUP BY equipo_evaluar
)


SELECT  estudiantes.usuario,
        estudiantes.equipo_pertenece,
        notas.final
FROM (
SELECT DISTINCT substring(equipo_pertenece,15) as equipo_pertenece,usuario from asistencia
) estudiantes 
LEFT JOIN notas ON (notas.equipo_evaluar=estudiantes.equipo_pertenece)
ORDER BY equipo_pertenece,usuario

""").show(100)
```
